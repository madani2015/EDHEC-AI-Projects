### Create baseline model using Ordinary Least Squares

```python
ols = LinearRegression()
ols.fit(X_train, y_train)

y_pred_ols = ols.predict(X_test)

mse_ols = mean_squared_error(y_test, y_pred_ols)
r2_ols = r2_score(y_test, y_pred_ols)

print("---- OLS Baseline ----")
print(f"Test MSE: {mse_ols:.3f}")
print(f"Test R^2: {r2_ols:.3f}")
```

### how to do ridge regresison 

```python
ridge = Ridge()
alpha_values = [0.001, 0.01, 0.1, 0.5, 1, 10, 100]

param_grid_ridge = {'alpha': alpha_values}
ridge_cv = GridSearchCV(ridge, param_grid_ridge, 
                        scoring='neg_mean_squared_error', 
                        cv=5, 
                        n_jobs=-1)
ridge_cv.fit(X_train, y_train)

print("---- Ridge Regression ----")
print("Best alpha:", ridge_cv.best_params_)
print("Best CV Score (negative MSE):", ridge_cv.best_score_)

# Evaluate the best model on the test set
best_ridge = ridge_cv.best_estimator_
y_pred_ridge = best_ridge.predict(X_test)

mse_ridge = mean_squared_error(y_test, y_pred_ridge)
r2_ridge = r2_score(y_test, y_pred_ridge)

print(f"Test MSE: {mse_ridge:.3f}")
print(f"Test R^2: {r2_ridge:.3f}")
```

### how to do Lasso regresison 

```python
lasso = Lasso()
alpha_values_lasso = [0.0001, 0.001, 0.01, 0.3, 0.5, 1, 10]

param_grid_lasso = {'alpha': alpha_values_lasso}
lasso_cv = GridSearchCV(lasso, param_grid_lasso,
                        scoring='neg_mean_squared_error',
                        cv=5,
                        n_jobs=-1)
lasso_cv.fit(X_train, y_train)

print("---- Lasso Regression ----")
print("Best alpha:", lasso_cv.best_params_)
print("Best CV Score (negative MSE):", lasso_cv.best_score_)

best_lasso = lasso_cv.best_estimator_
y_pred_lasso = best_lasso.predict(X_test)

mse_lasso = mean_squared_error(y_test, y_pred_lasso)
r2_lasso = r2_score(y_test, y_pred_lasso)

print(f"Test MSE: {mse_lasso:.3f}")
print(f"Test R^2: {r2_lasso:.3f}")

# Check how many coefficients are zero
coef_lasso = best_lasso.coef_
num_zero_coefs = np.sum(coef_lasso == 0)
print(f"Number of coefficients shrunk to zero: {num_zero_coefs}/{p}")
```

## Let's focus on Lasso and its cool feature of removing unimportant features

we select several values for alpha, and then we plot the number of nonzero features left, together with the mse

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score

# Initialize the Lasso model
lasso = Lasso(random_state=123)

# Define the alpha values to test
alpha_values_lasso = [0.0001, 0.001, 0.01, 0.3, 0.5, 1, 10]

# Create a grid of hyperparameters for cross-validation
param_grid_lasso = {'alpha': alpha_values_lasso}

# Perform GridSearchCV with 5-fold cross-validation
lasso_cv = GridSearchCV(lasso, param_grid_lasso,
                        scoring='neg_mean_squared_error',
                        cv=5,
                        n_jobs=-1)

# Fit the model to the training data
lasso_cv.fit(X_train, y_train)

# Extract the best hyperparameters and cross-validation score
print("---- Lasso Regression ----")
print("Best alpha:", lasso_cv.best_params_['alpha'])
print("Best CV Score (negative MSE):", lasso_cv.best_score_)

# Extract the best model and make predictions on the test set
best_lasso = lasso_cv.best_estimator_
y_pred_lasso = best_lasso.predict(X_test)

# Calculate metrics on the test set
mse_lasso = mean_squared_error(y_test, y_pred_lasso)
r2_lasso = r2_score(y_test, y_pred_lasso)

print(f"Test MSE: {mse_lasso:.3f}")
print(f"Test R^2: {r2_lasso:.3f}")

# Analyze coefficients
coef_lasso = best_lasso.coef_
num_zero_coefs = np.sum(coef_lasso == 0)
p = len(coef_lasso)  # Total number of features
print(f"Number of coefficients shrunk to zero: {num_zero_coefs}/{p}")

# Create a plot of MSE vs. Number of Non-Zero Coefficients
mse_test_values = []
num_nonzero_coefs = []

for alpha in alpha_values_lasso:
    lasso_temp = Lasso(alpha=alpha, random_state=123)
    lasso_temp.fit(X_train, y_train)
    y_pred_temp = lasso_temp.predict(X_test)
    mse_temp = mean_squared_error(y_test, y_pred_temp)
    mse_test_values.append(mse_temp)
    num_nonzero_coefs.append(np.sum(lasso_temp.coef_ != 0))

# Plot MSE and number of non-zero coefficients
plt.figure(figsize=(12, 6))

# Left plot: MSE vs. alpha
plt.subplot(1, 2, 1)
plt.plot(alpha_values_lasso, mse_test_values, marker='o', label='Test MSE')
plt.xscale('log')
plt.xlabel('Alpha (log scale)')
plt.ylabel('MSE')
plt.title('MSE vs Alpha')
plt.grid(True)
plt.legend()

# Right plot: Number of non-zero coefficients vs. alpha
plt.subplot(1, 2, 2)
plt.plot(alpha_values_lasso, num_nonzero_coefs, marker='o', label='Non-Zero Coefficients', color='orange')
plt.xscale('log')
plt.xlabel('Alpha (log scale)')
plt.ylabel('Number of Non-Zero Coefficients')
plt.title('Non-Zero Coefficients vs Alpha')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

# Plot y_test vs y_pred for the best model
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred_lasso, alpha=0.7, edgecolor='k')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--', label='Perfect Fit')
plt.title('y_test vs y_pred (Lasso Regression)')
plt.xlabel('Actual Values (y_test)')
plt.ylabel('Predicted Values (y_pred)')
plt.legend()
plt.grid(True)
plt.show()
```
