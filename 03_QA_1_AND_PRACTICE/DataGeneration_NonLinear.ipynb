```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

N= 20000
A = 10
B = 10
x =np.random.uniform(low=-10, high=15, size=N)


w = 5
eps = np.random.laplace(loc=0, scale=.2, size=N)

y = (1*np.sin(.2*x + (x/w)**3) + eps)*np.exp(-(x)/B)

plt.plot(x,y, '.')

data = pd.DataFrame({'x': x, 'y': y})
data.to_csv('non_linear_data.csv', index=False)
```
